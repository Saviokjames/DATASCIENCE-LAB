```python


x = df['sepal.length']

plt.hist(x, bins=30, color="red",alpha=0.8)
plt.title('HISTOGRAM')
plt.xlabel('Sepal Length')

plt.show()
```

```python
x=df['sepal.length']
y=df['sepal.width']
plt.scatter(x,y,color='blue')
plt.xlabel('length')
plt.ylabel('width')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x1=df['sepal.length']
y1=df['petal.length']
x2=df['sepal.width']
y2=df['petal.width']
plt.scatter(x2,y2,label='sepel',color='blue',marker='x',s=100)
plt.scatter(x2,y2,label='petel',color='yellow',marker='o')
plt.legend()
plt.title('SCATTER PLOT')
plt.show()
```

```python
sns.kdeplot(df,fill=True,color='green')
plt.show()
```

```python
sns.boxplot(df,color='green')
```

```python
x=pd.plotting.andrews_curves(df, 'variety')
x.plot()
plt.show()
```
