```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris=pd.read_csv('iris.csv')
```

```python
iris.hist()
plt.show()
```

```python
iris.plot(kind="box")
plt.show()
```

```python
x=iris['sepal.length']
y=iris['sepal.width']
plt.scatter(x,y,color='blue')
plt.xlabel('length')
plt.ylabel('width')
plt.title('SCATTER PLOT')
plt.show()
```

```python
sns.kdeplot(iris,fill=True,color='green')
plt.show()
```

```python
x=pd.plotting.andrews_curves(iris, 'variety')
x.plot()
plt.show()
```
